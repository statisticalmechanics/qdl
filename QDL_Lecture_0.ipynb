# Basics of Machine Learning

**Machine learning**

* supervised learning
  * regression
  * classification: logistic regression

* unsupervised learning
  * clustering
  * dimensionality reduction
* reinforcement learning

**Deep learning**

* artificial neural network: neuron, layer, activation
* feedforward neural network (FNN): single-layer and  multilayer perceptron (MLP)
* convolutional neural network (CNN)
* recurrent neural network (RNN)
* graph neural network (GNN)

**Concepts and algorithms of learning**

* datasets: training set, validation set, test set
* input and output
* error: training error, validation error
* bias-variance tradeoff
* underfitting and overfitting
* generalization
* loss function: MSE, cross entropy
* performance measure or metric: accuracy, precision, recall, ROC, AUC
* parameters (weights) and hyperparameters
* learning curve
* model selection
* regularization
* optimization: stochastic gradient descent (SGD), learning rate
* backpropagation
* epoch, iteration, batch
* activation function: logistic, ReLu
* data augmentation
* dimensionality (of data)
* complexity (of model)
